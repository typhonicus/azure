### Executing the required transformations

```python
new_df = customers.join(order,["Customer_Id"])
```

```python
display(new_df)
```

```python
import pyspark.sql.functions as F
from pyspark.sql.functions import count,mean,sum,min,max,datediff,to_date,lit
```

##### Calculating the number of orders per customer

```python
df2 = new_df.groupBy("Customer_ID").agg(count("order_id"))
df2 = df2.withColumnRenamed("count(order_id)","no_of_orders")
display(df2)
```

```python
df3 = order.join(transactions,["order_id"])
```

##### Calculating amount spent per customer

```python
d4 = df3.groupBy("Customer_ID").agg(sum("sales"))
d4 = d4.withColumnRenamed("sum(sales)","amnt_spent")
display(d4)
```

##### Calculating purchase frequency of each customer

```python
df5 = df3.groupBy("Customer_ID").agg(datediff(to_date(lit(max("order_purchase_date"))),to_date(lit(min("order_purchase_date")))))
df5 = df5.withColumnRenamed("datediff(to_date(max(order_purchase_date)), to_date(min(order_purchase_date)))","duration")
df6 = df5.join(df2,["Customer_ID"])
df6 = df6.withColumn("purchase_frequency", (F.col("duration") / F.col("no_of_orders")))
df6 = df6.drop("duration","no_of_orders")
display(df6)
```

##### Calculating the number of returns per customer

```python
df7 = order.join(returns,["order_id"])
```

```python
df8 = df7.groupBy("customer_id").agg(count("return_reason"))
```

```python
df8 = df8.withColumnRenamed("count(return_reason)", "no_of_returns")
display(df8)
```

```python
customers_fact = df2.join(d4, ["customer_id"])
display(customers_fact)
```

```python
customers_fact = customers_fact.join(df6, ["customer_id"])
```

```python
display(customers_fact)
```

### Joining to create the desired customers_fact table

```python
customers_fact = customers_fact.join(df8,["customer_id"])
```

```python
display(customers_fact)
```
